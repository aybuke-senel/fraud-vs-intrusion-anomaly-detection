# Cross-Domain Anomaly Detection Comparison

Financial Fraud vs Cyber Intrusion Detection

## Objective

Compare anomaly detection performance across two domains:

- Financial Fraud Detection
- Cyber Attack Detection

Goal:
Understand how the same anomaly detection approach behaves in different domains.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Model Performance Across Domains

```python
comparison = pd.DataFrame({
    "Domain": ["Fraud", "Cyber"],
    "Best Model": ["Isolation Forest", "Isolation Forest"],
    "Recall": [0.16, 0.8619]
})

comparison
```

```python
comparison.set_index("Domain")["Recall"].plot(kind="bar")

plt.title("Fraud vs Cyber Recall Comparison")
plt.ylabel("Recall")
plt.savefig("domain_recall_comparison.png")
plt.show()
```

## Analysis

The same anomaly detection model shows significantly different performance across domains.

Fraud Detection:
- Extremely imbalanced data
- Fraud behavior overlaps with normal transactions
- Hard to detect anomalies

Cyber Intrusion Detection:
- More structured and distinct patterns
- Clear separation between normal and attack traffic
- Easier anomaly detection

This explains the large difference in recall.

## Why Does Performance Differ?

Financial Fraud:
- Extremely imbalanced dataset
- Fraud patterns are subtle and overlap with normal transactions

Cyber Intrusion:
- More structured and distinct attack patterns
- Clear behavioral differences between normal and attack traffic

As a result, anomalies are easier to detect in cyber data.

## Model Behavior

Isolation Forest outperformed LOF in both domains.

This indicates that tree-based anomaly detection methods are more robust across different types of data.

## Nature of Anomalies

Fraud:
- Subtle deviations
- High overlap with normal data

Cyber:
- Strong pattern deviations
- More distinguishable anomalies

## Business Impact

Fraud:
- Missing fraud leads to financial loss
- High recall is critical

Cyber:
- Detecting attacks prevents system breaches
- Balanced precision and recall is important

## Final Conclusion

Anomaly detection is not domain-specific.

The same analytical approach can be applied to both financial fraud detection and cyber intrusion detection.

However, model performance varies depending on the structure and nature of the data.
